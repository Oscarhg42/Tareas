````julia
using PyPlot
````

````julia
x=collect(0.0:1.0/2^2:4)
````

````julia
length(x)
````

````julia
y=x.^2 #este punto y gorrito significa que cada elemento del arreglo se eleva al cuadrado
````

````julia
plot(x,y,"ro-")
````

````julia
plot(x,y,"g.-")
xlabel(L"x",fontsize=20)
ylabel(L"y=x^2",fontsize=20)
````
